```python
#ANSWER 2:
# Printing a table with labels like 1a, 1b, 1c ... 4a, 4b, 4c

# Outer loop for numbers
for i in range(1, 5):   # 1 to 4

    # Inner loop for letters
    for j in ['a', 'b', 'c']:
        print(f"{i}{j}", end="  ")
    print()
```

```python
#ANSWER 3:
# All possible outcomes of rolling 2 dice

for i in range(1, 7):       # First dice (1–6)
    for j in range(1, 7):   # Second dice (1–6)
        print(f"({i}, {j})",end="  ")
    print()
```

```python
#ANSWER 4:
# Alphabet Pyramid
n = 5
for i in range(n):
    ch = chr(65 + i)       # 65 = 'A'
    for j in range(i + 1):
        print(ch, end=" ")
    print()
```

```python
#ANSWER 4: SECOND TERM OF ASCII:
# Alphabet Pyramid

n = 4  # number of rows

for i in range(1, n + 1):
    for j in range(65, 65 + i):  # ASCII 65 = 'A'
        print(chr(j), end=" ")
    print()  # new line
```

```python
#ANSWER 5:
# Number Pyramid

n = 5
for i in range(1, n+1):
    for j in range(i):
        print(i, end=" ")
    print()
```

```python
#ANSWER 6:
# Reverse Number Pyramid

n = 4  # number of rows
for i in range(n):
    numbers = " ".join(str(j) for j in range(5, 5 - (n - i), -1))
    print(numbers.rjust(n*2))

```

```python
#ANSWER 10:
n = 4  # number of rows

ch = 65  # ASCII value of 'A'

for i in range(n):
    # print alphabets up to (2*i + 1) letters
    for j in range(2 * i + 1):
        print(chr(ch + j), end=" ")
    print()
```

```python
# ANSWER 11:
def pascal_triangle(n):
    for line in range(n):
        num = 1
        for i in range(line + 1):
            print(num, end=" ")
            num = num * (line - i) // (i + 1)
        print()

# print first 6 rows
pascal_triangle(6)
```

```python
#ANSWER 12:
def factorial(n):
    if n == 1:
        print("1", end="")
        return 1
    print(n, end=" * ")
    return n * factorial(n - 1)

# num = int(input("Enter a number: "))
# print(f"\n{num}! = ", end="")
x=4
result = factorial(x)
print(result)
```

```python
#ANSWER 13:
n = 5

for i in range(1, n + 1):
    # left side (increasing numbers)
    for j in range(i, n + 1):
        print(j, end=" ")

    # right side (decreasing numbers)
    for j in range(n - 1, i - 1, -1):
        print(j, end=" ")

    print()
```

```python
#TRIANGLE STATEMENT:
def triangle_type(a, b, c):
    # Check if valid triangle
    if a + b <= c or a + c <= b or b + c <= a:
        return "Not a valid triangle"

    # Equilateral
    if a == b == c:
        return "Equilateral Triangle"

    # Isosceles
    elif a == b or b == c or a == c:
        return "Isosceles Triangle"

    # Scalene
    else:
        return "Scalene Triangle"


# Example
a, b, c = map(int, input("Enter 3 sides of a triangle (e.g., 3 4 5): ").split())
print("Triangle type:", triangle_type(a, b, c))
```
